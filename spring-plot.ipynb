```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
```

```python
# y axis
spring_weights = np.genfromtxt('spring_weights.csv', delimiter=',')

# x axis
spring_extensions = np.genfromtxt('spring_extensions.csv', delimiter=',')
```

```python
# not sure why, but values at 0 are 
# being read as nans.. fix them
spring_weights[0] = 0.110853
spring_extensions[0] = 0
```

```python
area = 200
plt.figure(figsize=(20,10))
plot_extensions = plt.scatter(spring_extensions, spring_weights, s=area, marker="x", vmin=0, label=('Spring'))

plt.title('Cumulative downward force versus cumulative spring extension', fontsize=30)
plt.xlabel('Cumulative spring extension (meters)', fontsize=20)
plt.ylabel('Cumulative downward Force (N)', fontsize=20)

plt.ylim(bottom=0)
plt.xlim(left=0)

plt.tick_params(axis='x', labelsize=16)
plt.tick_params(axis='y', labelsize=16)

# line of best fit
x = spring_extensions
y = spring_weights
slope, intercept = np.polyfit(x, y, 1)
poly1d_fn = np.poly1d([slope, intercept]) 
plt.plot(x, poly1d_fn(x), color='orange', alpha=1.0, label=('Line of best fit'))

plt.legend(fontsize=20, loc='lower right')
plt.show()
```
