```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
```

```python
meters = np.array([.1, .2, .3, .4, .5])
period_lg = np.array([0.36, 0.81, 1.21, 1.44, 1.96])
period_sm = np.array([0.36, 0.81, 1.21, 1.69, 1.96])
```

```python
plt.figure(figsize=(20,10))

area = 200
plot_lg = plt.scatter(period_lg, meters, s=area, marker="x", vmin=0, label='Large Bob')
plot_sm = plt.scatter(period_sm, meters, s=area, marker="+", vmin=0, label='Small Bob')

plt.title('Length versus Period of pendulum', fontsize=30)
plt.xlabel('Period of Pendulum squared (seconds^2)', fontsize=20)
plt.ylabel('Length of Pendulum (meters)', fontsize=20)

plt.ylim(bottom=0)
plt.xlim(left=0)

plt.tick_params(axis='x', labelsize=16)
plt.tick_params(axis='y', labelsize=16)

# get avg of both series for best fit
period_avg = np.array([0])
for i in range(0, len(period_lg)):
    period_avg = np.append(period_avg, (period_lg[i] + period_sm[i]) / 2)
    
# line of best fit
y = np.insert(meters, 0, 0)
slope, intercept = np.polyfit(period_avg, y, 1)
poly1d_fn = np.poly1d([slope, intercept]) 
plt.plot(period_avg, poly1d_fn(period_avg), color='grey', alpha=0.3, label="Line of best fit")

plt.legend(fontsize=20, loc='lower right')
plt.show()
```
